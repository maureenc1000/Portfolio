```python
# import the required libraies
import pandas as pd              # dataframes
import numpy as np               # numerical computing
import matplotlib.pyplot as plt  # visualizations
import seaborn as sns            # visualizations
from openpyxl import load_workbook  # correct default styles issue with excel files
```

```python
# remove future warnings that don't affect how current code runs
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
# define the file paths for the 3 excel datasets
file_paths = {
    "most_popular": "most-popular-netflix.xlsx",
    "all_weeks_global": "all-weeks-global-netflix.xlsx",
    "all_weeks_countries": "all-weeks-countries-netflix.xlsx"
}
```

```python
# create a function to apply default styles to datasets
def apply_default_styles(file_path):
    wb = load_workbook(file_path)  # Load workbook
    wb.save(file_path)  # Save it again to apply default styles
```

```python
# apply apply_default_styles function to all files
for file in file_paths.values():
    apply_default_styles(file)
```

### Clean up the data

```python
# clean data proactively for further analysis to account for infinite values that may occur
most_popular_data.replace([float('inf'), -float('inf')], float('nan'), inplace=True)
all_weeks_global_data.replace([float('inf'), -float('inf')], float('nan'), inplace=True)
all_weeks_countries_data.replace([float('inf'), -float('inf')], float('nan'), inplace=True)
```

```python
# convert date columns to datetime format for time series plotting
all_weeks_global_data["week"] = pd.to_datetime(all_weeks_global_data["week"])
```

```python
# fill missing values based on results for each data set above
most_popular_data["season_title"].fillna("N/A", inplace=True)
all_weeks_global_data["season_title"].fillna("N/A", inplace=True)
all_weeks_countries_data["season_title"].fillna("N/A", inplace=True)
all_weeks_global_data["runtime"].fillna(0, inplace=True)
all_weeks_global_data["weekly_views"].fillna(0, inplace=True)
```

```python
# classify content type to Movie or TV
most_popular_data["content_type"] = most_popular_data["category"].apply(lambda x: "TV Show" if "TV" in x else "Movie")
all_weeks_global_data["content_type"] = all_weeks_global_data["category"].apply(lambda x: "TV Show" if "TV" in x else "Movie")
```

### Create Visualizations

```python
# set plot style for seaborn visualizations
sns.set_style("whitegrid")
```

```python
#Visual 1: Top Categories by Popularity Over Time (Movie vs Television)
plt.figure(figsize=(12, 6))
sns.lineplot(data=all_weeks_global_data, 
             x='week', 
             y='weekly_hours_viewed', 
             hue='content_type',
             estimator='sum', 
             ci=None)
plt.title('Weekly Hours Viewed Over Time: Movie vs Television')
plt.xlabel('Week')
plt.ylabel('Weekly Hours Viewed')
#plt.legend(title='Genre')
plt.show()
```

```python
#Visual 2: top 10 most watched titles by total hours
top_titles = (
    all_weeks_global_data.groupby('show_title')['weekly_hours_viewed']
    .sum()
    .nlargest(10)
    .index
)

# Filter dataset for only those top ten titles
df_top_titles = all_weeks_global_data[all_weeks_global_data['show_title'].isin(top_titles)]

# Plot weekly time series
plt.figure(figsize=(14, 7))
sns.lineplot(
    data=df_top_titles,
    x='week',
    y='weekly_hours_viewed',
    hue='show_title',
    estimator=None,
    ci=None
)

plt.title('Top 10 Most Watched Titles Globally Over Time')
plt.xlabel('Week')
plt.ylabel('Weekly Hours Viewed')
plt.legend(title='Show Title', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
#Visual 3: Top 10 Most Watched Television Titles Globally
# get top ten tv shows
top_tv = (
    all_weeks_global_data[all_weeks_global_data['content_type'] == 'TV Show']
    .groupby('show_title')['weekly_hours_viewed']
    .sum()
    .nlargest(10)
)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_tv.values, y=top_tv.index, palette='Blues')
plt.title('Top 10 Most Watched TV Shows Globally')
plt.xlabel('Total Hours Viewed')
plt.ylabel('TV Show Title')
plt.show()
```

```python
#Visual 4: Top 10 Most Watched Movie Titles Globally
# get top ten movies
top10_movies = (
    all_weeks_global_data[all_weeks_global_data['content_type'] == 'Movie']
    .groupby('show_title')['weekly_hours_viewed']
    .sum()
    .nlargest(10)
)
plt.figure(figsize=(12, 6))
sns.barplot(x=top10_movies.values, y=top10_movies.index, palette='coolwarm')
plt.title('Top 10 Most Watched Movie Titles')
plt.xlabel('Total Hours Viewed')
plt.ylabel('Show Title')
plt.show()
```

```python
#Visual 5:. Seasonal Trends in Viewing Habits
all_weeks_global_data['month'] = all_weeks_global_data['week'].dt.month
plt.figure(figsize=(12, 6))
sns.boxplot(x='month', y='weekly_hours_viewed', data=all_weeks_global_data)
plt.title('Seasonal Trends in Viewing Habits')
plt.xlabel('Month')
plt.ylabel('Weekly Hours Viewed')
plt.show()
```

```python
#Visual 6:. Movie vs. TV Show Performance
plt.figure(figsize=(8, 6))
sns.boxplot(x='content_type', y='weekly_hours_viewed', data=all_weeks_global_data, palette='pastel')
plt.title('Comparison of Movie vs. TV Show Performance')
plt.xlabel('Content Type')
plt.ylabel('Weekly Hours Viewed')
plt.show()
```
